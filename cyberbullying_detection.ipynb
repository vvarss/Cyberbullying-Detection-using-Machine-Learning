```python
sym = '!"#$%&\'()*+,-./:;<=>?@[\\]^_`{|}~'
```

```python
def clean(cell):
    for i in sym:
        cell = cell.replace(i," ") 
    cell = cell.replace("  "," ")   #double space replaced with single quotations
    cell = cell.replace("   "," ")  #triple space replaced with single quotation
    cell = cell.replace("    "," ") #quadraple space replaced with space quotation
    return cell
```

```python
tv = TfidfVectorizer(max_features=3000)
x2 = tv.fit_transform(df['tweet_text']).toarray()
```

```python
y = df['cyberbullying_type']
```

```python
x1_train,x1_test,y1_train,y1_test=train_test_split(x1,y,train_size=0.80,random_state=42)
x2_train,x2_test,y2_train,y2_test=train_test_split(x2,y,train_size=0.80,random_state=42)
```

```python
rf = RandomForestClassifier()
model1 = rf.fit(x2_train,y2_train)
model1.score(x2_test,y2_test)
```
